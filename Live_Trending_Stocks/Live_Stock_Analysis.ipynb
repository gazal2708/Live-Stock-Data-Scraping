```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
import lxml
```

```python
link = "https://www.investing.com/equities/trending-stocks"
```

```python
driver = webdriver.Chrome()
driver.get(link)
driver.implicitly_wait(100)
```

```python
from selenium.webdriver.common.action_chains import ActionChains
```

```python
highChartElements= driver.find_elements_by_xpath("//*[name()='div' and @id='trendingByPopularityChart']//*[name()='div' and @id='highcharts-0']//*[name()='svg']//*[name()='g' and @class='highcharts-series highcharts-tracker']/*[name()='rect']")
```

# Trending Stocks by Popularity

```python
for element in highChartElements:
    hov = ActionChains(driver).move_to_element(element)
    hov.perform()
    tip =  driver.find_elements_by_xpath("//*[name()='div' and @class='highcharts-tooltip']//*[name()='span']")
    for tipel in tip:
        print(tipel.text) 
```

# Trending stocks by sector

# Trending stocks quotes(price)

```python
tb =  driver.find_elements_by_xpath("//*[name()='a' and @id='filter_price']")
for x in tb:
    h = ActionChains(driver).move_to_element(x)
    h.click().perform()
    el = driver.find_elements_by_xpath("//*[name()='div' and @id='trendingInnerContent']//*[name()='table']")
    for y in el:
        print(y.text)
```
